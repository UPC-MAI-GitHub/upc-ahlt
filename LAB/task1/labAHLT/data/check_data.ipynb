```python
import pandas as pd

df = pd.read_csv('vals.txt',sep='|', header=None,names=['Entity','Type'])

df = df.drop_duplicates(subset='Entity',keep='first')

df.groupby('Type').count()
```

```python
df['len'] = df['Entity'].apply(lambda x: len(x))
```

```python
drugs = df[df['Type']=='drug']
drugs['suffix'] = drugs['Entity'].str[-4:]
drugs.head(100)
```

```python
brands = df[df['Type']=='group']

brands_uniq = brands['Entity'].drop_duplicates(keep='first')

sorted(brands_uniq,reverse=False)
```

```python
df['first_word_upper'] = df['Entity'].str[0]
df['is_upper'] = df['first_word_upper'].apply(lambda x: x.isupper())
```

```python
df[df['is_upper']==True].groupby('Type').count()
```

```python
k = df[df['is_upper']==True]
sorted(k[k['Type'] == 'group'].Entity.unique())
```

```python
df['leng'] = df['Entity'].apply(lambda x: len(x))
```

```python
groups = df[df['Type']=='group'].drop_duplicates(subset='Entity',keep='first').sort_values('leng')
```

```python
groups.sort_values('leng',ascending=False)
```

```python
kk = df.sort_values('leng',ascending=False)
aa = kk[kk['Type'] == 'drug_n'].drop_duplicates('Entity',keep='first')
aa[aa['leng'] > 4].Entity.values
```
